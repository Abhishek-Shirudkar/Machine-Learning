```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("./canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (USD)')
plt.scatter(df.year, df['per capita income (USD)'], color='red', marker='+')
```

Now Linear Regression model.

```python
new_df_without_income = df.drop('per capita income (USD)', axis= 'columns')
new_df_without_income
```

```python
model = linear_model.LinearRegression()
model.fit(new_df_without_income, df['per capita income (USD)'])
```

```python
p = model.predict([[2025]])
p
```

```python
model.coef_
```

```python
model.intercept_
```
